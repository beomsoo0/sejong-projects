```
import tensorflow as tf

tf.__version__
```

```
# 상수
a = tf.constant(1.0)
b = tf.constant(2.0)

c = a + b

print(c.numpy())
print(c)
```

```
# 변수
W = tf.Variable(tf.random.normal([1]))
print("initial W=", W.numpy())
print("==========================================")

# session 필요 x, numpy() 메서드 이용하여 numpy 값 리턴

for step in range(2):
  W = W + 1.0
  print('step =', step, "W =", W.numpy())
```

```
# 함수
a = tf.constant(1.0)
b = tf.constant(3.0)

def tensor_sum(x, y):
  return x + y

result = tensor_sum(a, b)

print(type(result))
print(type(result.numpy()))
print(result.numpy())
```
