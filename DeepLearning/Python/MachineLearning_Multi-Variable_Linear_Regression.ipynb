```python
import numpy as np

# Training Data [25 x 3]

loaded_data = np.loadtxt("C:/Users/bsbs7/Desktop/py_temp/data-01-test-score.csv",delimiter=',', dtype=np.float32)

x_data = loaded_data[:, 0:-1]
t_data = loaded_data[:, [-1]]
```

```python
# 임의의 직선 y = Wx + b 정의, 가중치 W와 바이어스 b 임의의 값으로 초기화
W = np.random.rand(3,1)
b = np.random.rand(1)
print("W =", W, ", W,shape =", W.shape, ", b =", b, ", b.shape =", b.shape)
```

```python
# 손실함수
def loss_func(x, t):
    y = np.dot(x,W) + b
    
    return ((np.sum((t - y) ** 2)) / len(x))
```

```python
# 수치미분 함수
def numerical_derivative(f, x):
    delta_x = 1e-4 # 0.000
    grad = np.zeros_like(x) # x 크기의 어레이 선언 및 초기화
    
    it = np.nditer(x, flags = ['multi_index'], op_flags = ['readwrite'])
    
    while not it.finished:
        idx = it.multi_index
        tmp_val = x[idx] # x 바뀌기 때문에 저장해둠
        x[idx]= float(tmp_val) + delta_x
        fx1 = f(x) # f(x+delta_x)
        
        x[idx] = tmp_val - delta_x
        fx2 = f(x) # f(x-delta_x)
        grad[idx] = (fx1 - fx2) / (2*delta_x)
        
        x[idx] = tmp_val
        it.iternext()
        
    return grad
```

```python
# 손실함수 값 계산 함수
# 입력변수 x, t : numpy type
def error_val(x, t):
    y = np.dot(x, W) + b
    
    return( np.sum( (t-y)**2 )) / len(x)

# 학습을 마친 후, 임의의 데이터에 대해 미래 값 예측 함수
#입력변수 x : numpy type
def predict(x):
    y = np.dot(x, W) + b
    
    return y
```

```python
learning_rate = 1e-3 # 발산하는 경우 1e-3 ~ 1e-6 등으로 바꿔 실행

f = lambda x : loss_func(x_data, t_data) # f(x) = loss_func(x_data, t_data)

print("Initial error value =", error_val(x_data, t_data), "Initial W =", W, "\n", ", b=", b)

for step in range(400001):
    
    W -= learning_rate * numerical_derivative(f, W)
    b -= learning_rate * numerical_derivative(f, b)
    
    if (step % 400 == 0):
        print("step =", step, "error value =", error_val(x_data, t_data), "W =", W, ", b=", b)
    
```
